```python
import cv2

A1_img = cv2.imread('../datasets/A1.jpg')
B1_img = cv2.imread('../datasets/B1.jpg')
plt.imshow(A1_img)
plt.show()
plt.imshow(B1_img)
plt.show()
```

```python
# テンプレート画像の幅と高さを取得
wA, hA, _ = A1_img.shape
wB, hB, _ = B1_img.shape

# 画像の検索（Template Matching）
result = cv2.matchTemplate(A1_img, B1_img, cv2.TM_CCORR_NORMED)
# 検索結果の信頼度と位置座標の取得
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
```

```python
# 検索結果の左上頂点の座標を取得
top_left = max_loc

# 検索結果の右下頂点の座標を取得
bottom_right = (top_left[0] + wA, top_left[1] + hA)

# 検索対象画像内に、検索結果を長方形で描画
cv2.rectangle(A1_img, top_left, bottom_right, (255, 255, 0), 2)

# 画像を表示
cv2.imshow('img', A1_img)
```
